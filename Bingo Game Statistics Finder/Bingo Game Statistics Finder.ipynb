```python
## Create random Bingo cards and .
import random
from random import randrange
from random import shuffle
NUMS_PER_LETTER = 15

## Create a Bingo card with randomly generated numbers
# @return a dictionary representing the card where the keys are the strings ‘‘B’’, ‘‘I’’, ‘‘N’’,
# ‘‘G’’, and ‘‘O’’, and the values are lists of the numbers that appear under each letter
# from top to bottom

def createCard():
    card = {}
    
    # The range of integers that can be generated for the current letter
    lower = 1
    upper = 1 + NUMS_PER_LETTER
    
    # For each of the five letters
    for letter in ["B", "I", "N", "G", "O"]:
        # Start with an empty list for the letter
        card[letter] = []

        # Keep generating random numbers until we have 5 unique ones
        while len(card[letter]) != 5:
            next_num = randrange(lower, upper)
            # Ensure that we do not include any duplicate numbers
            if next_num not in card[letter]:
                card[letter].append(next_num)
        
        # Update the range of values that will be generated for the next letter
        lower = lower + NUMS_PER_LETTER
        upper = upper + NUMS_PER_LETTER
        if letter == 'N':
            card[letter][2] = 'X'
  
    # Return the card
    return card

## Display a Bingo card with nice formatting
# @param card the Bingo card to display
# @return (None)

def displayCard(card):
    # Display the headings
    print("    B     I     N     G     O")
    # Display the numbers on the card
    for i in range(5):
        for letter in ["B", "I", "N", "G", "O"]:
            print("%5s " % card[letter][i], end="")
        print()

## Create cards, check each card and calculate average number of drawings to win
# @return the last win card

def play():
    # Read number of cards from palyer to be generated and checked
    n = int(input('Enter the number of games to be played: '))
    # Create list of numbers form 1 to 75 and shuffle them in order to name drawings
    list1 = list(range(1, 76))
    random.shuffle(list1)
    # Create an empty list and COUNT variable in order to be able to collect results
    list2 = []
    COUNT = 0
    
    for i in range(n):
        card = createCard()
        win = False
        for a in list1:
            for i in range(5):
                for letter in ["B", "I", 'N', "G", "O"]:
                    # Replace numbers from cards with 'X' as the number from your card and drawings are the same
                    if card[letter][i] == a:
                        card[letter][i] = 'X'
                    # Wining cases
                    if card["B"][0] == 'X' and card["I"][1] == 'X' and card["G"][3] == 'X' and card["O"][4] == 'X':
                        win = True
                    if card["O"][0] == 'X' and card["G"][1] == 'X' and card["I"][3] == 'X' and card["B"][4] == 'X':
                        win = True
                    if card["B"][0] == 'X' and card["O"][4] == 'X' and card["B"][4] == 'X' and card["O"][0] == 'X':
                        win = True
                    if card['B'][0] == 'X' and card['B'][1] == 'X' and card['B'][2] == 'X' and card['B'][3] == 'X' and card['B'][4] == 'X':
                        win = True
                    if card['I'][0] == 'X' and card['I'][1] == 'X' and card['I'][2] == 'X' and card['I'][3] == 'X' and card['I'][4] == 'X':
                        win = True
                    if card['N'][0] == 'X' and card['N'][1] == 'X' and card['N'][3] == 'X' and card['N'][4] == 'X':
                        win = True
                    if card['G'][0] == 'X' and card['G'][1] == 'X' and card['G'][2] == 'X' and card['G'][3] == 'X' and card['G'][4] == 'X':
                        win = True
                    if card['O'][0] == 'X' and card['O'][1] == 'X' and card['O'][2] == 'X' and card['O'][3] == 'X' and card['O'][4] == 'X':
                        win = True
                    if card['B'][0] == 'X' and card['I'][0] == 'X' and card['N'][0] == 'X' and card['G'][0] == 'X' and card['O'][0] == 'X':
                        win = True
                    if card['B'][1] == 'X' and card['I'][1] == 'X' and card['N'][1] == 'X' and card['G'][1] == 'X' and card['O'][1] == 'X':
                        win = True
                    if card['B'][2] == 'X' and card['I'][2] == 'X' and card['G'][2] == 'X' and card['O'][2] == 'X':
                        win = True
                    if card['B'][3] == 'X' and card['I'][3] == 'X' and card['N'][3] == 'X' and card['G'][3] == 'X' and card['O'][3] == 'X':
                        win = True
                    if card['B'][4] == 'X' and card['I'][4] == 'X' and card['N'][4] == 'X' and card['G'][4] == 'X' and card['O'][4] == 'X':
                        win = True
            COUNT += 1               
            if win:
                break
        list2.append(COUNT)
        COUNT = 0
    # Display the result    
    print('Average balls that should be drawn to win out of 75: ', sum(list2)/ len(list2))
    print('During', n, 'bingo games, the maximum amount of drawings to win was: ', max(list2))
    print('During', n, 'bingo games, the minimum amount of drawings to win was: ', min(list2))
    print()
    # Display the last card
    print('The last win card: ')
    return displayCard(card)

# Call the play function
play()         
```
